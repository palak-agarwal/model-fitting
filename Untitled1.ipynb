```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('shampoo.csv',header=0)
df=df[:-1]
df.head()
```

```python
df['year']=df.apply(lambda row:row.Month,axis=1)
df.head()
```

```python
df['year']=df.apply(lambda row:row.year[0],axis=1)
df.head()
```

```python
df['Month']=df.apply(lambda row:row.Month[2:],axis=1)
df.head()
```

```python
df['Month']=df.apply(lambda row:int(row.Month),axis=1)
df.head()
```

```python
dummy=pd.get_dummies(df['year'])
dummy.head()
```

```python
dummy=pd.concat([dummy,df],axis=1)
dummy.head()
```

```python
df=dummy.drop('year',axis=1)
df.head()
```

```python
dummy=pd.get_dummies(df['Month'])
dummy.head()
```

```python
dummy=pd.concat([dummy,df],axis=1)
dummy.head()
```

```python
df=dummy.drop('1',axis=1)
df.head()
```

```python
df=dummy.drop(12,axis=1)
df.head()
#df.keys()
```

```python
df=dummy.drop('Month',axis=1)
df.head()
```

```python
x = df.iloc[:, :-1]
x.head()
```

```python
y = df.iloc[: , -1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size= 0.25)  #we removed random state then accuracy improved
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
a = regressor.score(X_test, Y_test)
```
