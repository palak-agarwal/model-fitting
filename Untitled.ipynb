```python
dummy=pd.get_dummies(df['year'])
dummy
```

```python
dummy=pd.concat([dummy,df],axis=1)
dummy
```

```python
df=dummy.drop('year',axis=1)
df
```

```python
x = df.iloc[:, :-1]
x.head()
```

```python
y = df.iloc[: , -1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size= 0.25)  #we removed random state then accuracy improved
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
a = regressor.score(X_test, Y_test)
```
