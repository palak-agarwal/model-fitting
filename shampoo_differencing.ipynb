```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('shampoo.csv',header=0,index_col=0)
df=df[:-1]
df.head()
```

```python
df.plot()
plt.show()
```

```python
#de-trend by differencing
x=df.values
diff=list()
for i in range(1, len(x)):
    value = x[i]-x[i-1]
    diff.append(value)
plt.plot(diff)
plt.show()
```
